# SQuAD JSON Extraction Pipeline

This notebook extracts valid questions and answers from a JSON file (SQuAD format) and filters them to create a clean CSV for evaluation.

### Requirements:
- Extract `question` and `text` (under "answers").
- Only include where `"is_impossible": false`.
- Select 100 random questions.
- Save to `data/extracted_math_questions.csv`.


```python
import json
import pandas as pd
import random

# File paths
input_path = 'data/train-v2.0.json'
output_path = 'data/dataset_general_knowledge.csv'

# Load JSON
with open(input_path, 'r', encoding='utf-8') as f:
    squad_data = json.load(f)

extracted_data = []

# SQuAD format extraction logic
for entry in squad_data['data']:
    for paragraph in entry['paragraphs']:
        for qa in paragraph['qas']:
            # Only extract answers with is_impossible: false
            if not qa.get('is_impossible', True):
                question_text = qa.get('question', '')
                # Taking the first answer if it exists
                answers_list = qa.get('answers', [])
                if answers_list:
                    answer_text = answers_list[0].get('text', '')
                    extracted_data.append({
                        'question': question_text,
                        'answer': answer_text
                    })

# Convert to DataFrame
df = pd.DataFrame(extracted_data)

# Limit to 100 random questions
if len(df) > 100:
    df_sampled = df.sample(n=100, random_state=42)
else:
    df_sampled = df

# Save to CSV
df_sampled.to_csv(output_path, index=False)

print(f"Total valid entries found: {len(df)}")
print(f"Saved {len(df_sampled)} random questions to: {output_path}")
df_sampled.head()
```
